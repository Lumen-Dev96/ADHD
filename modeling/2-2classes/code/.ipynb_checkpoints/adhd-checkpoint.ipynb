```python
from bokeh.plotting import figure, show, curdoc, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.layouts import layout, gridplot, column
from bokeh.io import push_notebook
import numpy as np
from datetime import datetime
from random import randint
import sys


# 初始化數據
data_length = 30

data_channels = 12

x = np.arange(data_length)

y_data = [np.zeros(data_length) for _ in range(data_channels)]

grapId = 0

# 初始化 Bokeh 的 ColumnDataSource
source1 = ColumnDataSource(data=dict(x=x, y=y_data[0]))
source2 = ColumnDataSource(data=dict(x=x, y=y_data[1]))
source3 = ColumnDataSource(data=dict(x=x, y=y_data[2]))

source4 = ColumnDataSource(data=dict(x=x, y=y_data[3]))
source5 = ColumnDataSource(data=dict(x=x, y=y_data[4]))
source6 = ColumnDataSource(data=dict(x=x, y=y_data[5]))

source7 = ColumnDataSource(data=dict(x=x, y=y_data[6]))
source8 = ColumnDataSource(data=dict(x=x, y=y_data[7]))
source9 = ColumnDataSource(data=dict(x=x, y=y_data[8]))

source10 = ColumnDataSource(data=dict(x=x, y=y_data[9]))
source11 = ColumnDataSource(data=dict(x=x, y=y_data[10]))
source12 = ColumnDataSource(data=dict(x=x, y=y_data[11]))


# 創建 Bokeh 的線型圖
plots = []
plot1 = figure(plot_height=300, title="MPU1", tools="crosshair,pan,reset,save,wheel_zoom")
line = plot1.line('x', 'y', source=source1, line_width=2, line_alpha=0.8, line_color='black')
line = plot1.line('x', 'y', source=source2, line_width=2, line_alpha=0.8, line_color='blue')
line = plot1.line('x', 'y', source=source3, line_width=2, line_alpha=0.8, line_color='red')
line = plot1.line('x', 'y', source=source4, line_width=2, line_alpha=0.8, line_color='green')
line = plot1.line('x', 'y', source=source5, line_width=2, line_alpha=0.8, line_color='yellow')
line = plot1.line('x', 'y', source=source6, line_width=2, line_alpha=0.8, line_color='pink')



# plot2 = figure(plot_height=300, title="Channel AY1", tools="crosshair,pan,reset,save,wheel_zoom")
# line = plot2.line('x', 'y', source=source2, line_width=2, line_alpha=0.8)

# plot3 = figure(plot_height=300, title="Channel AZ1", tools="crosshair,pan,reset,save,wheel_zoom")
# line = plot3.line('x', 'y', source=source3, line_width=2, line_alpha=0.8)

# plot4 = figure(plot_height=300, title="Channel GX1", tools="crosshair,pan,reset,save,wheel_zoom")
# line = plot4.line('x', 'y', source=source4, line_width=2, line_alpha=0.8)

# plot5 = figure(plot_height=300, title="Channel GY1", tools="crosshair,pan,reset,save,wheel_zoom")
# line = plot5.line('x', 'y', source=source5, line_width=2, line_alpha=0.8)

# plot6 = figure(plot_height=300, title="Channel GZ1", tools="crosshair,pan,reset,save,wheel_zoom")
# line = plot6.line('x', 'y', source=source6, line_width=2, line_alpha=0.8)

plot7 = figure(plot_height=300, title="MPU2", tools="crosshair,pan,reset,save,wheel_zoom")
line = plot7.line('x', 'y', source=source7, line_width=2, line_alpha=0.8)
line = plot7.line('x', 'y', source=source8, line_width=2, line_alpha=0.8)
line = plot7.line('x', 'y', source=source9, line_width=2, line_alpha=0.8)
line = plot7.line('x', 'y', source=source10, line_width=2, line_alpha=0.8)
line = plot7.line('x', 'y', source=source11, line_width=2, line_alpha=0.8)
line = plot7.line('x', 'y', source=source12, line_width=2, line_alpha=0.8)


# plot8 = figure(plot_height=300, title="Channel AY2", tools="crosshair,pan,reset,save,wheel_zoom")
# line = plot8.line('x', 'y', source=source8, line_width=2, line_alpha=0.8)

# plot9 = figure(plot_height=300, title="Channel AZ2", tools="crosshair,pan,reset,save,wheel_zoom")
# line = plot9.line('x', 'y', source=source9, line_width=2, line_alpha=0.8)

# plot10 = figure(plot_height=300, title="Channel GX2", tools="crosshair,pan,reset,save,wheel_zoom")
# line = plot10.line('x', 'y', source=source10, line_width=2, line_alpha=0.8)

# plot11 = figure(plot_height=300, title="Channel GY2", tools="crosshair,pan,reset,save,wheel_zoom")
# line = plot11.line('x', 'y', source=source11, line_width=2, line_alpha=0.8)

# plot12 = figure(plot_height=300, title="Channel GZ2", tools="crosshair,pan,reset,save,wheel_zoom")
# line = plot12.line('x', 'y', source=source12, line_width=2, line_alpha=0.8)

plots.append(plot1)
plots.append(plot7)
# plots.append(plot3)
# plots.append(plot4)
# plots.append(plot5)
# plots.append(plot6)
# plots.append(plot7)
# plots.append(plot8)
# plots.append(plot9)
# plots.append(plot10)
# plots.append(plot11)
# plots.append(plot12)


grid = layout(plots, sizing_mode='scale_width',)

# 將繪圖添加到文檔
#curdoc().add_root(layout([[plot1]], sizing_mode='scale_width'))


# 更新函數，每次呼叫時更新數據
def update(message):
    global x, y_data, grapId
    
    # 模擬獲取新數據，這裡使用 randint 生成隨機數
    new_data = np.random.randint(-10, 10, size=data_length)
    
    payload = message.payload.decode()
    data = json.loads(payload)
    mqData = np.array(data)
    channel1 = mqData[:, 0]
    channel2 = mqData[:, 1]
    channel3 = mqData[:, 2]
    channel4 = mqData[:, 3]
    channel5 = mqData[:, 4]
    channel6 = mqData[:, 5]
    channel7 = mqData[:, 6]
    channel8 = mqData[:, 7]
    channel9 = mqData[:, 8]
    channel10 = mqData[:, 9]
    channel11 = mqData[:, 10]
    channel12 = mqData[:, 11]
    
    

    print('mqttdata:', mqData.size)
    
    # 更新數據源的數據
    grapId += data_length
    source1.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel1))
    source2.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel2))
    source3.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel3))
    source4.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel4))
    source5.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel5))
    source6.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel6))
    source7.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel7))
    source8.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel8))
    source9.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel9))
    source10.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel10))
    source11.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel11))
    source12.stream(dict(x=np.arange(grapId, grapId + data_length), y=channel12))
    


    push_notebook()

# 在Jupyter Notebook中顯示
output_notebook()

show(grid, notebook_handle=True, notebook_url='localhost:8888')












import paho.mqtt.client as mqtt
import json

def on_connect(client, userdata, flags, rc):
    if rc == 0:
        print("Connected to MQTT broker")
    else:
        print("Failed to connect, return code %d\n", rc)

def on_message(client, userdata, message):
    update(message)
    print("Received data from esp32")
    #print("Received message on topic %s: %s" % (message.topic, message.payload.decode()))

# 连接MQTT代理
broker_address = "broker.emqx.io"
port = 1883
username = "test"
password = "123456"

client = mqtt.Client()
client.username_pw_set(username, password)
client.on_connect = on_connect
client.on_message = on_message


client.connect(broker_address, port)


# 订阅消息
topic = "Lumen"
client.subscribe(topic)

client.loop_forever()
```
